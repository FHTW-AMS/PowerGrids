## Version control with Git

 Version control systems like git are used in software development to store code, version it, and enable collaboration.Individual granular changes are recorded and saved with a message. Advantages:

* Automatic backup of your code
* Older versions can be restored at any time
* Traceability: Why changes were introduced, when a new problem occurs, who developed that part of the code
* Easy code sharing via open-source platforms like github.com
* Collaboration: Many developers can work on the same code. Advanced concepts like branches and merges help keep changes separate and then merge them back together.

### Setting up your own Git repository
*  Install git: https://git-scm.com/downloads.

* Create a user account at https://github.com/signup.

* Create a repository with a name, description, default Readme.md and optionally private.

* Follow the setup steps on GitHub.

### Minimal command line example using the existing repository

The following commands are executed in the command line; for example, open Git Bash (or PowerShell).
You should see this once, but most development environments also provide a graphical interface to perform the steps.

To test the commands locally, it's enough to have git installed; to upload changes to GitHub, you need a user (see setup below).

1. Download the repository from github.com. This creates a new folder that can be moved at any time because all files for version control are stored directly in it.
    ```bash
    git clone https://github.com/FHTW-AMS/PowerGrids.git
    ```

2.  Make changes to one or more files. Review the changes since the last tracked version.
    ```bash
    # Change Directory to downloaded folder
    cd PowerGrids
    # Open Readme.md, do some changes, save and close
    git status
    git diff Readme.md
    ```

3. Add the changes to the next commit, record the commit with a short message.
   ```bash
    git add Readme.md
    git status
    git commit -m "Example Update of Readme.md"
    ```

4. Inspect the recent changes. Each commit consists of an ID, a message, the author and the time. It also shows which pointers (branches, remotes) reference a commit. With the commit ID you can restore old versions (`git checkout`) or compare changes and files.
    ```bash
    git log
    ```

5. Synchronize with the online repository: first fetch new updates and then upload local changes.
    ```bash
    git pull  # Get newest changes from repository
    git push
    ```

### Final remarks

The simplest commands for daily use should be known, but to truly exploit git's potential, you need a bit more.

For example, when you want to return to an old version or view specific changes. The IDE integration is often the easiest option here; for the command line, the help commands (`git help` or `git <command> help`), the internet, or us can help.
A small warning when using notebooks: because notebooks are not a pure text format, merging changes works relatively poorly. There are fewer problems if you clear outputs before committing, but it is generally recommended that only one person makes changes to a notebook. If you make your own experiments in this repository and want to track them in git, it's better to save them as a new notebook to avoid incompatible changes on the next pull.

## Python installation and packages with conda

In general, any recent Python installation should work; the notebooks may use some functionality that is only included from around 3.10. For a new installation I recommend miniconda, which manages different Python versions and packages for different projects well. Here is a list of steps to install miniconda:

* Download miniconda: https://docs.conda.io/projects/miniconda/en/latest/miniconda-install.html

* Start Anaconda Prompt; to make conda available in other terminals, run the following command:
  ```bash
  conda init
  ```

* In the terminal of your choice, create a new conda environment named fh_intro:
  ```bash
  conda create -n fh_intro python>=3.11
  ```
  
* Activate the environment (this step must always be executed; in the development environment you can often set the environment for a project)
  ```bash
  conda activate fh_intro
  ```

All packages used in this repository are listed in *requirements.txt*; install them with

  ```bash
  pip install -r requirements.txt
  ```
