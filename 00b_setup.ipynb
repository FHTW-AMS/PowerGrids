# Setup Python environment - step by step

This notebook is meant to be a step by step guide for setting up a Python development environment.
An environment represents a specific collection of Python packages and their versions. This is important to ensure that the code runs correctly, as different versions of packages can have different functionalities or dependencies. By creating a separate environment for each project, you can avoid conflicts between packages and maintain a clean and organized workspace. There are different tools to create and manage Python environments, such as virtualenv, pipenv, or conda. In this guide, we will use conda, which is a popular package manager that also allows you to create and manage environments easily.

### Set up Python environment

The FH Pandapower Cosimulation repository and the included notebooks use some libraries that are not included in Anaconda.
Therefore, we create a new conda environment in which we install all packages. We use the Anaconda Navigator for this,
although the command line could also be used.

* First, we add the conda_forge channel. Channels > Add... > **conda-forge**
* Then we import the environment from the [environment.yaml](environment.yaml)
* Alternatively, this can also be done in the CLI:
    `conda create --file environment.yaml -n name`
